集成学习只是模型融合的一部分，并非等价。

集成学习就是把一系列弱的学习器集合成强学习器。

集成学习主要包括三方面：bagging（bootstrap），boosting（叠加式算法），stacking（模型1的结果作为特征输入到模型2中）。

bagging,通过降低方差的形式降低测试误差。

随机森林：bagging+特征采样
特征采样：比如从十个特征中取出5个特征。



Bagging主要通过降低方差的方式减少预测误差。
Boosting思想提高最终的预测效果是通过不断减少偏差的形式，与Bagging有着本质的区别。

强可学习和弱可学习的概念：
  - 弱学习：识别错误率小于1/2（即准确率仅比随机猜测略高的学习算法）                     
  - 强学习：识别准确率很高并能在多项式时间内完成的学习算法  
 强可学习难以获得，弱可学习比较容易，在PCA框架下，两者是等价的（？ AdaBoost算法是为了证明弱可学习和强可学习算法等价而提出的

随机森林与bagging的联系与区别：
随机森林的思想仍然是bagging，但是它选择使用CART决策树作为弱学习器，以及在此基础上在节点上所有的n个样本特征中选择一个最优的特征来做决策树的左右子树划分。

什么是bagging：
bagging（bootstrap），就是随机采样构建样本，通过这些样本训练弱模型，再根据集合策略形成强学习模型.

什么是bootstrap：
自助采样(bootstrap)，即有放回的从数据集中进行采样，也就是说，同样的一个样本可能被多次进行采样。
